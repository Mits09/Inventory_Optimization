```python
date_cols = ['Received_Date','Last_Purchase_Date','Expiry_Date','Audit_Date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format="mixed", dayfirst=True, errors="coerce")

```

```python
date_cols
```

```python
money_cols = ['Unit_Cost_USD','Last_Purchase_Price_USD','Total_Inventory_Value_USD']

for col in money_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['Avg_Daily_Sales'] = df['Avg_Daily_Sales'].astype(str).str.replace(',', '.').astype(float)
df['Days_of_Inventory'] = df['Days_of_Inventory'].astype(str).str.replace(',', '.').astype(float)
```

```python
df.isnull().sum()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.duplicated(['SKU_ID','Batch_ID']).sum()
```

```python
df['Logic'] = df['Quantity_On_Hand'] >= (df['Quantity_Reserved'] + df['Quantity_Committed'])
df
```

```python
import pandas as pd

# Load the cleaned dataframe. This file contains data after initial cleaning and encoding.
# This ensures 'df' is defined even if previous cells were not run or runtime was reset.
df = pd.read_csv("cleaned_inventory.csv")

# Re-apply transformations that occurred after saving to "cleaned_inventory.csv"
# These transformations ensure the columns used in ABC analysis are correctly formatted.

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Re-process percentage and comma-separated numeric columns
pct_cols = [
    'Supplier_OnTime_Pct',
    'Audit_Variance_Pct',
    'Demand_Forecast_Accuracy_Pct'
]
for col in pct_cols:
    # Ensure the column is string type before operations
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace('%','')
    df[col] = df[col].str.replace(',','.')
    df[col] = pd.to_numeric(df[col], errors='coerce')

comma_cols = [
    'Avg_Daily_Sales',
    'Days_of_Inventory'
]
for col in comma_cols:
    # Ensure the column is string type before operations
    df[col] = df[col].astype(str).str.replace(',','.')
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill any remaining NaN values with the median for numeric columns
df.fillna(df.median(numeric_only=True), inplace=True)


# Original code from DbczXZf9APSO to perform ABC analysis
abc_analysis = df.groupby('ABC_Class')['Total_Inventory_Value_USD'].sum().reset_index()

abc_analysis['Percentage'] = (
    abc_analysis['Total_Inventory_Value_USD'] /
    abc_analysis['Total_Inventory_Value_USD'].sum()
) * 100

abc_analysis.sort_values(by='Total_Inventory_Value_USD', ascending=False)
```
